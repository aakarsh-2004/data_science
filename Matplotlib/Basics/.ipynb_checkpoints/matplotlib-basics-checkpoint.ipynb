# Introduction to Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [1, 2, 3, 4, 10] # X axis
y = [4, 5, 6, 7, 8] # Y axis
plt.plot(x, y);
```

```python
# 1st method
fig = plt.figure() # create a figure
ax = fig.add_subplot() # add some axes
plt.show()
```

```python
# 2nd Method
fig = plt.figure() # creates a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot([10, 12, 23, 50, 100], [50, 100, 150, 200, 250]); # add some data
```

## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3 ,4]
y = [11, 22, 33 ,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) #(width, height)

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save and show (you save the whole figure)
fig.savefig('../../graph_1.png')
```

## Making figures with numpy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Sub plot

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use the same data and plot scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot 
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond Butter": 10,
                    "Peanut Butter": 8,
                    "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title= "Dan's Nut butter shop", ylabel="Price ($)");
```

```python
# Horizontal bar
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()

ax.hist(x);
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize=(10,5))

# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

# Plotting from pandas dataframes

```python
import pandas as pd
```

```python
# Make a dataframe
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000), 
               index= pd.date_range("1/1/2020", periods= 1000))
ts = ts.cumsum()
ts.plot()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
type(car_sales["Price"])
```

```python
# Remove last two zeroes
car_sales["Price"] = car_sales["Price"].str[0:-2]
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2023", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Let's plot the total sales
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
# Reassign price column to int
car_sales["Price"] = car_sales["Price"].astype(int)

# Plot scatter plot with price column as numeric
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
```
